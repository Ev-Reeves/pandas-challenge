```python
import pandas as pd
```

```python
schools_data_path = "Resources/schools_complete.csv"
students_data_path = "Resources/students_complete.csv"
```

```python
school_data = pd.read_csv(schools_data_path)
student_data = pd.read_csv(students_data_path)
```

```python
#District Summary
district_data = pd.merge(school_data, student_data, how = 'left', on = ['school_name', 'school_name'])
```

```python
district_data
```

```python
N_schools = district_data["School ID"].max()
N_students = len(district_data)
district_budget = str(district_data["budget"].sum())
mean_math_score = str(district_data["math_score"].mean())
mean_reading_score = str(district_data["reading_score"].mean())
```

```python
bins = [0,69,100]
bin_names = ["Fail", "Pass"]
```

```python
district_data["Math Score Summary"] = pd.cut(district_data["math_score"], bins, labels = bin_names, include_lowest = True)
district_data["Reading Score Summary"] = pd.cut(district_data["reading_score"], bins, labels = bin_names, include_lowest = True)
```

```python
math_pass_prop = (len(district_data.loc[district_data["Math Score Summary"]=="Pass"]))/(len(district_data))
math_pass = str(math_pass_prop*100)
reading_pass_prop = (len(district_data.loc[district_data["Reading Score Summary"]=="Pass"]))/(len(district_data))
reading_pass = str(reading_pass_prop*100)
```

```python
count_pass = district_data.groupby(['Math Score Summary', 'Reading Score Summary']).size()
count_pass
pass_fail = count_pass.to_frame(name = 'size').reset_index()
pass_fail['Overall'] = pass_fail['Math Score Summary'].astype(str) + pass_fail['Reading Score Summary'].astype(str) 
PassPass = pass_fail.loc[pass_fail['Overall']=='PassPass', 'size'].iloc[0]
both_pass_prop = PassPass/len(district_data)
both_pass = str(both_pass_prop*100)
```

```python
columns = ['Number of Schools', 'Number of Students', 'Total Budget', 'Average Math Score', 'Average Reading Score',
          'Percentage Passing Math', 'Percentage Passing English', 'Percentage of Students Passing Both']

sum_dict = {
    'Number of Schools': [N_schools],
    'Number of Students': [N_students],
    'Total Budget': ["$"+ district_budget],
    'Average Math Score': [mean_math_score+ "%"],
    'Average Reading Score': [mean_reading_score+ "%"],
    'Percentage Passing Math': [math_pass + "%"],
    'Percentage Passing English': [reading_pass + "%"],
    'Percentage of Students Passing Both': [both_pass + "%"]
}

district_summary = pd.DataFrame(sum_dict, columns=columns)
district_summary
```
